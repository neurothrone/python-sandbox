```python
import pandas as pd
import polars as pl
import seaborn as sns

# Load the dataset
diamonds = sns.load_dataset("diamonds")
```

```python
# Create a Pandas DataFrame
df_pd = pd.DataFrame(diamonds)
df_pd
```

```python
# Create a Polars DataFrame
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/diamonds.csv"
df_pl = pl.read_csv(url)
df_pl
```

```python
# Subset the Pandas DataFrame
df_pd[["cut", "price", "clarity"]]
```

```python
# Subset the Polars DataFrame
df_pl[["cut", "price", "clarity"]]
```

```python
# Use select() on the Polars DataFrame
df_pl.select(["cut", "price", "clarity"])
```

```python
# Filter the Pandas DataFrame
df_pd[df_pd["price"] > 1000]
```

```python
# Filter the Polars DataFrame
df_pl.filter(df_pl["price"] > 1000)
```

```python
# Sort the Pandas DataFrame
df_pd.sort_values(by="price", ascending=False)
```

```python
# Sort the Polars DataFrame
df_pl.sort("price")
```

```python
# Find the price for the 5th record in the dataset
print(df_pl[4, 6])
df_pl[4, "price"]
```
